```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('music_dataset.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
print("\nJumlah missing values di setiap kolom:")
print(df.isnull().sum())
```

```python
df_clean = df.copy()
```

```python
missing_values = df_clean.isnull().sum()
```

```python
duplicates = df_clean.duplicated().sum()
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
summary_stats = df_clean.describe()
```

```python
correlation = df_clean.corr(numeric_only=True)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Music Features")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_genres = df_clean.groupby("Genre")["Streams"].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette="viridis")
plt.title("Top 10 Genres by Average Streams")
plt.xlabel("Average Streams")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```
